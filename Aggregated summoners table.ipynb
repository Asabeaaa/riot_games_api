## Summing stats

```python
stats_df = new_summoner_df[['summonerId','kills','deaths','assists','totalDamageDealt','goldEarned','totalMinionsKilled','magicDamageDealt','physicalDamageDealt','longestTimeSpentLiving']]
```

```python
stats_df.head()
```

```python
stats_dff=stats_df.groupby('summonerId').sum().reset_index()
```

## Champion used

```python
champion_df = new_summoner_df[['summonerId','championId']]
```

```python
champion_df['Number_of_times_selected'] = 1
```

```python
champion_df.head()
```

```python
df_1 = champion_df.groupby(['summonerId', 'championId'])\
                  .agg({'Number_of_times_selected': 'sum'})\
                  .reset_index()
```

```python
df_1['championId'] = 'champion_' + df_1['championId'].astype(str)
```

```python
df_1.head()
```

```python
champion_dff=df_1.pivot(index='summonerId', columns='championId', values='Number_of_times_selected')
```

```python
champion_dff=champion_dff.reset_index()
```

```python
champion_dff.head()
```

```python
champion_dff.info()
```

## Calculating wins and losses

```python
wins_df = new_summoner_df[['summonerId','win']]
```

```python
wins_df.head()
```

```python
wins_df["count"] = wins_df["win"].astype(int)
```

```python
wins_df.head()
```

```python
df_2 = wins_df.groupby(['summonerId', 'win'])\
                  .agg({'count': 'sum'})\
                  .reset_index()
```

```python
df_2['win'] = df_2['win'].replace(True, 'Wins')
```

```python
df_2['win'] = df_2['win'].replace(False, 'Losses')
```

```python
df_2.head()
```

```python
wins_dff = df_2.pivot(index='summonerId', columns='win', values='count')
```

```python
wins_dff = wins_dff.reset_index()
```

```python
wins_dff.head()
```

```python
wins_dff.info()
```
